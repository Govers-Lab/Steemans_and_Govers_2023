## Load necessary libraries

```python
import pandas as pd
import pickle
import numpy as np
import glob
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_curve, auc
```

## Load data
If the labeled_svm_features pickle files from multiple image folders are not yet grouped together, relocate them in the same file. It is a necessity that the directory_path leads to the location of the files. Also note that you can specify the suffix of the file in 'pkl_files' (i.e. with '*labeled_svm_features.pkl', you will open the pkl-files ending on 'labeled_svm_features').

```python
# Specify the path to the folder of the labeled svm data pickle files
directory_path = 'C:/Users/u0158103/Documents/Labeled_Data/'

# Get a list of all pickle files in the directory. 
pkl_files = glob.glob(directory_path + '*labeled_svm_features.pkl')

# Initialize an empty list to store the loaded DataFrames
dataframes = []

# Load the DataFrames and specify how many cell are in each DataFrame
for pkl_file in pkl_files:
    with open(pkl_file, 'rb') as f:
        svm_dataset = pickle.load(f)
        svm_training = pd.DataFrame(svm_dataset)
        num_cells = len(svm_training)
        print(f"Loaded file: {pkl_file}")
        print(f"Number of cells in DataFrame: {num_cells}\n")
        dataframes.append(svm_training)

# Concatenate all the DataFrames into a single DataFrame
concatenated_df = pd.concat(dataframes, ignore_index=True)
```

## Data preparation
### Remove NaNs

```python
svm_training = concatenated_df.copy()
svm_training = svm_training[svm_training['discard'] != True]
svm_training = svm_training.dropna()

# Print the number of removed rows
num_removed_rows = len(concatenated_df) - len(svm_training)
print(f"Number of removed rows: {num_removed_rows}")
```

### Split data randomly in test- and training set
The data split can affect the performance of the model. In our case, a random split with a test-train ratio of 20-80 performed adequately.
### Check class proportions to detect class imbalance
Proportions of 'Good' and 'Bad' cells generally seem to be balanced. However, if a class has a ratio of 1-100 or lower, other methods have to be applied. Dealing with the class imbalance problem is touched upon in the troubleshooting section.

```python
# Extract the feature columns, excluding the first four columns (non-features)
feature_columns = svm_training.columns[4:-1]

## Split the dataset into training and testing sets
# Depending on the size of the dataset you can do a different split (i.e. test_size = 0.2) to increase the amount of data
# used in during the training of the SVM
train_data, test_data, train_labels, test_labels = train_test_split(
    svm_training[feature_columns], svm_training.iloc[:, -1], test_size=0.5, random_state=42)

## Calculate class proportions
# If there is a minority class (ratio 1-100) a different approach might be required (e.g. assigning weights 
# to the minority class, or downsampling majority class, ...)
labels_series = pd.Series(test_labels)
proportions = labels_series.value_counts(normalize=True)
print(proportions)

# Show first 5 rows of the dataset
svm_training.head()
```

### Check feature correlations to detect redundancies
Highly correlated features could potentially be removed without affecting the performance of the model. However, other approaches that eliminate features based on the performance of the model such as recursive feature elimination (RFE) might be more appropriate. Note that RFE does not work with the RBF kernel. SVMs utilizing the RBF kernel deal well with high dimensional data and feature selection matters less here. 

```python
correlations = svm_training[feature_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlations, cmap="coolwarm")
plt.colorbar(heatmap)
plt.title("Correlation Heatmap of Features")
plt.xticks(np.arange(0.5, len(correlations.columns), 1), correlations.columns, rotation=90)
plt.yticks(np.arange(0.5, len(correlations.index), 1), correlations.index)
plt.show()
```

## Train SVM with NystrÃ¶m method
The NystrÃ¶m method is a powerful technique used to approximate kernel functions, which plays a key role in reducing the computational complexity associated with solving optimization problems. By leveraging a subset of representative points from the dataset, the NystrÃ¶m method constructs a low-rank approximation of the kernel matrix. This approximation captures the essential information of the original kernel matrix while significantly reducing the number of computations required. The NystrÃ¶m method is particularly beneficial when dealing with large datasets, as it enables efficient and scalable solutions without sacrificing much accuracy. We often solely use this technique instead of the normal training procedure.

```python
from sklearn.kernel_approximation import Nystroem

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('nystroem', Nystroem(kernel='rbf')),
    ('svm', SVC())
])
parameters = {
    'nystroem__gamma': [1e-2],  # Gamma parameter for the NystrÃ¶m approximation
    'nystroem__n_components': [100],  # Number of components for the NystrÃ¶m approximation
    'svm__kernel': ['rbf'], # Add 'linear' to this array if you also want to test the linear kernel
    'svm__C': [100, 1000, 10000],
    'svm__gamma': [1e-3, 1e-2, 1e-1]
}

# Define the grid search object with 5-fold cross-validation
grid_search = GridSearchCV(pipeline, parameters, cv=5)

# Fit the grid search object to the training data
grid_search.fit(train_data, train_labels)

# Print the best hyperparameters and the corresponding cross-validation score
print("Best hyperparameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)

best_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('nystroem', Nystroem(kernel='rbf',
                          gamma=grid_search.best_params_['nystroem__gamma'],
                          n_components=grid_search.best_params_['nystroem__n_components'])),
    ('svm', SVC(kernel=grid_search.best_params_['svm__kernel'], 
                C=grid_search.best_params_['svm__C'], 
                gamma=grid_search.best_params_['svm__gamma'], probability=True))
])

# Train the best pipeline on the training data
best_pipeline.fit(train_data, train_labels)

accuracy = best_pipeline.score(test_data, test_labels)
print("Accuracy on test data:", accuracy)
```

## Train SVM without NystrÃ¶m method (not recommended for datasets exceeding 5000 cells)
In order to train the SVM using the optimal parameters k-fold cross-validation is performed with 2 different kernels ('linear' and 'rbf'), a range of C values (100, 1000, 10000) and a range of gamma values (1e-3, 1e-2, 1e-1). If none of the models seem to be performing well, the first solution might be to choose a different range of these values. Regarding the choice of kernel, the RBF-kernel performs well as the datapoint seem to be best separable in non-linear fashion. For larger datasets it is recommended to use the NystrÃ¶m method.

```python
# Define pipeline with a standard scaler and an SVM classifier to tune the hyperparameters
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Step 1: Feature scaling using StandardScaler
    ('svm', SVC())  # Step 2: Support Vector Machine (SVM) classifier
])

# Define the hyperparameters to tune
parameters = {
    'svm__kernel': ['linear', 'rbf'],  # SVM kernel types to try: linear and radial basis function (rbf)
    'svm__C': [100, 1000, 10000],  # Values of the SVM regularization parameter C to try
    'svm__gamma': [1e-3, 1e-2, 1e-1]  # Values of the SVM kernel coefficient gamma to try
}

# Define the grid search object with 5-fold cross-validation
grid_search = GridSearchCV(pipeline, parameters, cv=5)

# Fit the grid search object to the training data
grid_search.fit(train_data, train_labels)

# Print the best hyperparameters and the corresponding cross-validation score
print("Best hyperparameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)

# Define the pipeline with a standard scaler and an SVM classifier with the best hyperparameters
best_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Feature scaling using StandardScaler
    ('svm', SVC(kernel=grid_search.best_params_['svm__kernel'], 
                C=grid_search.best_params_['svm__C'], 
                gamma=grid_search.best_params_['svm__gamma'], probability=False))  # SVM classifier with best hyperparameters
])

# Train the best pipeline on the training data
best_pipeline.fit(train_data, train_labels)

# Calculate accuracy on the test data
accuracy = best_pipeline.score(test_data, test_labels)
print("Accuracy on test data:", accuracy)
```

## Validation with ROC-curve
For validation with an ROC-curve the probability argument in Pipeline() has to be 'True'. This will increase the computation time.
The AUC (Area Under the Curve) quantifies the model's ability to correctly rank instances from the positive and negative classes. A perfect classifier would achieve an AUC value of 1, indicating that it can perfectly distinguish between the two classes. On the other hand, a random classifier would have an AUC of 0.5, as it performs no better than random guessing.

```python
# Predict the probabilities of the test data
test_probabilities = best_pipeline.predict_proba(test_data)[:,1]

# Compute the false positive rate, true positive rate, and area under the ROC curve
fpr, tpr, thresholds = roc_curve(test_labels, test_probabilities)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(4,4))
plt.plot(fpr, tpr, color='purple', lw=2, label='ROC curve (AUC = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='g', lw=2, linestyle='--')
plt.xlim([0.0, 1.01])
plt.ylim([0.0, 1.01])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right", frameon = False)
plt.show()
```

## Confusion matrix
The confusion matrix is a tabular representation that summarizes the performance of a classification model by showing the counts of true positive (TP), true negative (TN), false positive (FP), and false negative (FN) predictions. Special attention has to be paid to the false positives as these wrongly classified cells will be further analyzed. It is important to minimize these cases and further investigate whether they are truly unfit for downstream analyses. 

```python
from sklearn.metrics import confusion_matrix
predicted_labels = best_pipeline.predict(test_data)

# Calculate the confusion matrix
cm = confusion_matrix(test_labels, predicted_labels)

# Define the labels for TP, TN, FP, FN
labels = np.array([['TN', 'FP'], ['FN', 'TP']])

# Create a figure and axis
fig, ax = plt.subplots()

# Create a heatmap using imshow
im = ax.imshow(cm, interpolation='nearest', cmap='Blues')

# Add colorbar

# Add labels for TP, TN, FP, FN
thresh = cm.max() / 2
for i in range(len(labels)):
    for j in range(len(labels)):
        color = 'white' if cm[i, j] > thresh else 'black'
        ax.text(j, i, f"{labels[i, j]}\n{cm[i, j]}", ha='center', va='center', color=color)

# Set labels and title
# Remove ticks and labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])

# Remove spines (borders) around the plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_title('Confusion Matrix')

# Adjust the layout
plt.tight_layout()

# Display the confusion matrix
plt.show()
# Extract the values from the confusion matrix
true_negative = cm[0, 0]
false_positive = cm[0, 1]
false_negative = cm[1, 0]
true_positive = cm[1, 1]

# Calculate accuracy
accuracy = (true_positive + true_negative) / np.sum(cm)

# Calculate precision
precision = true_positive / (true_positive + false_positive)

# Calculate recall (sensitivity)
recall = true_positive / (true_positive + false_negative)

# Calculate specificity
specificity = true_negative / (true_negative + false_positive)

# Calculate F1 score
f1_score = 2 * (precision * recall) / (precision + recall)

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Specificity:", specificity)
print("F1 Score:", f1_score)
```

## Evaluation of feature bias
By examining the division of the average feature values of the true labels (test_data_label1) and the predicted labels (predicted_labels_label1) for the samples with label 1, you can gain insights into potential bias in the predicted features. Deviations from 1 suggest that the classifier may be biased in its predictions for label 1 instances, indicating a potential bias in the classifier's overall performance.

```python
# Filter the test_data based on label 1
test_data_label1 = test_data[test_labels == 1]
# Calculate the average of each feature in test_data_label1
test_data_label1_average = test_data_label1.mean()
# Filter the predicted labels based on label 1
predicted_labels_label1 = predicted_labels[test_labels == 1]
# Calculate the average of each feature in predicted_labels_label1
predicted_labels_label1_average = test_data_label1[predicted_labels_label1 == 1].mean()
# Divide the average of each feature in test_data_label1 by the average of each feature in predicted_labels_label1
feature_division = test_data_label1_average / predicted_labels_label1_average
# Select the first twelve columns from the feature_division series (morphological characteristics)
selected_features = feature_division.index[:12]
selected_divisions = feature_division[:12]

plt.figure(figsize=(5, 5))  # Set the figure size
# Create a scatter plot
plt.axhline(y=1, color='red', linestyle='--', linewidth=2)
plt.scatter(selected_features, selected_divisions, marker='o', s=50, color='orange', edgecolor='black')

plt.ylabel('Visually curated score / SVM-curated score')
plt.xticks(rotation=90)
plt.ylim(0.94, 1.06)  # Set the y-axis limits

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Set thicker linewidth for bottom and left spines
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)

# Show only the bottom and left ticks
plt.gca().xaxis.set_ticks_position('bottom')
plt.gca().yaxis.set_ticks_position('left')

plt.tight_layout()
plt.show()
```
